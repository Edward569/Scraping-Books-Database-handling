#  Write a sql query to get books that has rate >=3 and has 'Mr' in its name.

```python
df = pd.read_sql("SELECT * FROM Books WHERE stars >=3 OR prices > 20 ", connection)
df
```

# Write a sql query to get books that has rate not 3.

```python
df = pd.read_sql("SELECT * FROM Books WHERE stars !=3", connection)
df
```

# Write a sql query to get books that has price between 10 and 40 and has rate 3 or 4 or 1.

# Write a sql query to count the number of books that have 'Secret' in thier names and price between 10£ and 25£.

```python
df = pd.read_sql("SELECT COUNT(titles) AS COUNTBooks  FROM Books WHERE titles ='Secret' AND prices BETWEEN 10 AND 25 ", connection)
df
```

# Write a sql query to get the minimum & maximum price for all the books that have rate 5.

```python
df = pd.read_sql("SELECT MIN(prices) AS SmallestPrices FROM Books WHERE stars = 5 ", connection)
df
```

# Write a sql query to calculate the avg price for all the books that have rate 5

```python
df = pd.read_sql("SELECT AVG(prices) AS averagetPrices FROM Books WHERE stars = 5 ", connection)
df
```

# Write a sql query to sum all book's price that have rate 2 and price between 10 and 40
